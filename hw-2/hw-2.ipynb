# Тема “Теоремы о пределах”

### 1. Найти предел

 - **a.** $\lim_{x \to 0} \frac{sin(2x)}{4x} = \lim_{x \to 0} \frac{sin(2\cdot0)}{4\cdot0} = \left [ \frac{0}{0} \right ]$

 $\lim_{x \to 0} \frac{sin(2x)}{4x} = \lim_{x \to 0} \frac{\mathbf{sin(2x)}\cdot2}{\mathbf{2x}\cdot4} = \lim_{x \to 0} \frac{2}{4} = \frac{1}{2}$
 
 **$\ast$ первый замечательный предел**

 **Ответ:** $\lim_{x \to 0} \frac{sin(2x)}{4x} = \frac{1}{2}$

 - **b.** $\lim_{x \to 0} \frac{x}{sin(x)} = \left [ \frac{0}{0} \right ]$

 $\lim_{x \to 0} \frac{\mathbf{x}}{\mathbf{sin(x)}} = 1$
 
 **$\ast$ первый замечательный предел**

 **Ответ:** $\lim_{x \to 0} \frac{x}{sin(x)} = 1$

 - **c.** $\lim_{x \to 0} \frac{x}{arcsin(x)} = \left [ \frac{0}{0} \right ]$

 $\lim_{x \to 0} \frac{x}{arcsin(x)} = \lim_{x \to 0} \frac{sin(x)}{sin(arcsin(x))} = \lim_{x \to 0} \frac{\mathbf{sin(x)}}{\mathbf{x}} = 1$
 
 **$\ast$ первый замечательный предел**

 **Ответ:** $\lim_{x \to 0} \frac{x}{arcsin(x)} = 1$

 - **d.** $\lim_{x \to \infty} \left (\frac{4x+3}{4x-3} \right )^{6x} = \left [ \left(\frac{\infty}{\infty} \right)^{\infty} \right ]$

 $\lim_{x \to \infty} \left (\frac{4x+3}{4x-3} \right )^{6x} =$
 
 $= \lim_{x \to \infty} \left (\frac{4x-3+6}{4x-3} \right )^{6x} =$
 
 $= \lim_{x \to \infty} \left (\frac{4x-3}{4x-3} + \frac{6}{4x-3} \right )^{6x} =$
 
 $= \lim_{x \to \infty} \left (1 + \frac{6}{4x-3} \right )^{6x} =$
 
 $= \lim_{x \to \infty} \left (\mathbf{1 + \frac{6}{4x-3}} \right )^{\mathbf{\frac{4x-6}{6}}\cdot\frac{6}{4x-3}\cdot6x} =$

 $= e^{\lim_{x \to \infty} \frac{36x}{4x-3}} =$
 
 $= e^{\lim_{x \to \infty} \frac{\frac{36x}{x}}{\frac{4x}{x}-\frac{3}{x}}} =$
 
 $= e^{\frac{36}{4}} = e^{9}$

 **$\ast$ второй замечательный предел**

 **Ответ:** $\lim_{x \to \infty} \left (\frac{4x+3}{4x-3} \right )^{6x} = e^{9}$
